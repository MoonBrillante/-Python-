安装成功后在终端输入scrapy 检查版本

startproject: Creates a new projet创建一个新项目
    
genspider: Generates a new spider from a template根据模板生成一个新爬虫
    
crawl:Runs a spider 执行爬虫
    
shell:Starts the interactive scraping console 启动交互式抓取控制台


## 2. 创建一个项目
$scrapy sartproject example

$cd example


items.py： 该文件定义了待抓取域的模型。

settings.py： 该文件定义了一些设置， 如用户代理、爬取延时等。

spiders/： 该目录存储实际的爬虫代码。

另外， Scrapy 使用scrapy.cfg 设置项目配置，使用pipelines.p y
处理要抓取的域。
